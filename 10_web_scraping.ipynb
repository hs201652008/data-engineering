```python
"""
Exercise 1: Search '메타버스' news on Naver
"""
import requests
url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%A9%94%ED%83%80%EB%B2%84%EC%8A%A4'

res = requests.get(url)

html_doc = res.text
print(html_doc)
```

```python
"""
Use BeautifulSoup Parser
"""
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
# Search all the <a> tags
soup.findAll('a')
```

```python
# Extract news information

news = soup.findAll("a", {"class":"news_tit"})
```

```python
news
```

```python
# First news
news[0]
```

```python
# Get a url of the first news
news[0].get('href')
```

```python
# Get a title
news[0].get('title')
```

```python
len(news)
```

```python
for i in range(len(news)):
    print(news[i].get('title'))
```
